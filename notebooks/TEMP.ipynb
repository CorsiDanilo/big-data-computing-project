```python
bs_all_gbt = pd.read_csv('/content/bs_all_gbt.CSV')
```

```python
bs_all = pd.concat([bs_all_lr, bs_all_glr, bs_all_rf, bs_all_gbt])
bs_all
```

Percentage of correct predictions for ('LinearRegression', 'final_trained'): 44.65%

Percentage of correct predictions for ('GeneralizedLinearRegression', 'final_trained'): 43.82%

Percentage of correct predictions for ('RandomForestRegressor', 'final_trained'): 45.33%

Percentage of correct predictions for ('GradientBoostingTreeRegressor', 'final_trained'): 45.55%

```python
import numpy as np
from google.colab import autoviz

def violin_plot(df, value_colname, facet_colname, figscale=1, mpl_palette_name='Dark2', **kwargs):
  from matplotlib import pyplot as plt
  import seaborn as sns
  figsize = (12 * figscale, 1.2 * figscale * len(df[facet_colname].unique()))
  plt.figure(figsize=figsize)
  sns.violinplot(df, x=value_colname, y=facet_colname, palette=mpl_palette_name, **kwargs)
  sns.despine(top=True, right=True, bottom=True, left=True)
  return autoviz.MplChart.from_current_mpl_state()

chart = violin_plot(bs_all, *['RMSE', 'Features'], **{'inner': 'stick'})
chart
```
